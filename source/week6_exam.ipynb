## **[예제1]**

```python
# [실습 10]--------------------------------------
# 연속 if

 # 사용자가 입력한 점수(score)가 90 이상이면 A, 그렇지 않고 80이상이면 B, 그렇지 않고 70 이상이면 C, 그렇지 않고 60 이상이면 D, 그렇지 않으면 F

score = int(input("점수를 입력하시오:"))

# 만약 score가 90 이상이면 - A
# 그렇지 않고 만약 score가 80 이상이면 - B
# 그렇지 않고 만약 score가 70 이상이면 - C
# 그렇지 않고 만약 score가 60 이상이면 - D
# 그렇지 않으면 - F

if score >= 90 :
    if score>=95 :
        print('A+')
    else :
        print('A')
elif score >= 80 :
    print('B')
elif score >= 70 :
    print('C')
elif score >= 60 :
    print('D')
else :
    print('F')

print('end!')
```

## **[예제2]**
 가위, 바위, 보 게임을 통해서 컴퓨터와 대결해 봅시다.
* 잘 알겠지만, 가위바위보 게임은 다음의 규칙을 따릅니다. 나를 기준으로 컴퓨터가 내는 것으로 확인하면 됩니다.

```python
# [실습 12]--------------------------------------
# 임의의 값 2

# my_hand vs com_hand

import random

# 사용자의 가위/바위/보 중 하나를 입력받아 myHand 변수에 저장
my_hand = input('나의 가위/바위/보 ==> ')

# 컴퓨터의 가위/바위/보는 랜덤하게 추출하여 comHand에 저장
com_hand = random.choice(['가위', '바위', '보'])

# 게임 진행하기 위해 사용자의 가위/바위/보에 따라 if문을 사용함
# my_hand가 가위 이면 - 가위라고 출력
if my_hand == '가위' :
    print('가위')
# my_hand가 바위 이면 - 바위라고 출력
if my_hand == '바위' :
    print('바위')
# my_hand가 보 이면 - 보라고 출력
if my_hand == '보' :
    print('보')
```

```python
# 사용자의 가위/바위/보 중 하나를 입력받아 myHand 변수에 저장
my_hand = input('나의 가위/바위/보 ==> ')

# 컴퓨터의 가위/바위/보는 랜덤하게 추출하여 comHand에 저장
com_hand = random.choice(['가위', '바위', '보'])

print(f'{com_hand=}')
# my_hand가 가위 이면 -
    ## com_hand가 가위 이면 비김
if my_hand == '가위':
    if com_hand == '가위':
        print('비겼습니다')
    elif com_hand == '바위' :
        print('졌습니다')
    else :
        print('이겼습니다')
    ## com_hand가 바위 이면 "졌습니다" 출력

    ## com_hand가 보 이면 "이겼습니다" 출력

# my_hand가 바위 이면 - 바위라고 출력
elif my_hand == '바위' :
    if com_hand == '바위' :
        print('비겼습니다')
    elif com_hand == '보' :
        print('졌습니다')
    else :
        print('이겼습니다')
# my_hand가 보 이면 - 보라고 출력
elif my_hand == '보' :
    if com_hand == '보' :
        print('비겼습니다')
    elif com_hand == '가위' :
        print('졌습니다')
    else :
        print('이겼습니다')
else :
    print('잘못된 입력입니다. 가위 바위 보 중에 하나를 내시오')
```

* 나와 컴퓨터가 각각 1부터 6 사이의 주사위를 굴립니다. 숫자가 더 큰 쪽이 승리하며, 같을 경우 무승부로 판정하는 프로그램을 만들어 보세요.
  
* 목표:   난수 생성, 비교 연산자(>=, <) , 연속 if(if ~elif~else) 활용 마스터하기

-----------------------------------<br>
결과 : 무승부입니다! 다시 도전해 보세요.<br>


----
**[미션 가이드]**
➊  import random을 작성하여 랜덤 기능을 불러옵니다.<br>
➋  내 주사위(my_dice)와 컴퓨터 주사위(com_dice) 변수를 만들고, random.randint(1, 6)을 사용해 각각 1~6 사이의 랜덤한 숫자를 저장합니다.<br>
➌   print()를 사용해 "나의 주사위: [숫자], 컴퓨터의 주사위: [숫자]"를 화면에 보여줍니다.<br>
➍    if ~ elif ~ else 문을 사용하여 내 숫자가 클 때(result변수에 '승리' 저장), 컴퓨터가 클 때(result 변수에 '패배' 저장), 같을 때(result 변수에 '무승부'저장')의 결과를 출력하세요.<br>
[주의] 변수이름은, my_dice, com_dice, 그리고 결과는 result로 해주기 바랍니다.


```python
# [미션 2]--------------------------------------
# 주사위 배틀 게임

import random

# 내 주사위와 컴퓨터 주사위 굴리기
my_dice = random.randint(1, 6)
com_dice = random.randint(1, 6)

# 주사위 결과 출력
print(f"나의 주사위 : {my_dice}")
print(f"컴퓨터의 주사위 : {com_dice}")
print("-----------------------------------")

# 승패 판정
if my_dice > com_dice:
    result = '승리'
    print(f"결과 : 당신의 {result}입니다! 축하합니다! ^^")
elif my_dice < com_dice:
    result = '패배'
    print(f"결과 : 컴퓨터의 {result}입니다! 아쉽네요 ㅠㅠ")
else:
    result = '무승부'
    print(f"결과 : {result}입니다! 다시 도전해 보세요.")
```

```python
# [최종 점검] ------------------------------------------------
# 아래 코드를 실행하여 모든 테스트를 통과해야 합니다.
# 에러(AssertionError)가 발생하면 위 실습 코드를 다시 확인하세요.

try:
    # 1. 변수 존재 여부 체크
    assert 'my_dice' in globals(), "실습 1: my_dice 변수가 생성되지 않았습니다."
    assert 'com_dice' in globals(), "실습 2: com_dice 변수가 생성되지 않았습니다."
    assert 'result' in globals(), "실습 2: result 변수가 생성되지 않았습니다."

    # 2. 값의 유효성 체크
    assert 1 <= my_dice <= 6, "my_dice 값이 주사위 범위(1~6)를 벗어났습니다."
    assert result in ["승리", "패배", "무승부"], "result에 올바른 판정 결과가 없습니다."

    # 3. 모든 관문 통과 시 출력
    print("✨ 모든 실습 미션 통과!")
    print("PASS77")

except AssertionError as e:
    print(f"❌ 미션 실패: {e}")
```

## **[자율미션 2]-one more!!**
키와 몸무게를 입력하면 키에 맞는 표준 체중을 구한 후, 자신의 몸무게와 표준 체중과의 차이를 기준으로 저체중, 표준, 과체중 중에 하나를 출력하는 프로그램 작성


*   bmi = (자신의 키 - 100) * 0.9
* '표준’: (표준체중 – 3kg) 이상이면서  (표준체중 + 3kg) 이하인 체중
* '저체중’: (표준체중 – 3kg)보다 작은 체중
* '과체중': (표준체중 + 3kg)보다 큰 체중





```python
# 키(height) 사용자 입력
height = int(input("키를 입력하시오: "))
# 몸무게(weight) 사용자 입력
weight = int(input("몸무게를 입력하시오:"))
# bmi 계산
bmi = (height-100)*0.9
# 연속 if
if weight >= bmi-3 and weight <= bmi+3 :
    print("표준 체중입니다.")
elif weight < bmi-3 :
    print("저체중입니다.")
else :
    print("과체중입니다.")
```
