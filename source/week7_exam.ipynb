# **2부. for...range 문**





 ## **2.1 range 객체**
 - range(start=0, stop, steps=1)
 - start에서 stop-1 까지 연속된 정수 생성

```python
# [실습 6]--------------------------------------
range(0, 5, 1)
range(0, 5)   # step  생략시 기본은 1
range(5)   # start 생략시 기본은 0
print(range(5))


# range 연습 - 0에서 4까지 1씩 증가하는 수 생성
r1 = list(range(0,5,1))
print(f'{r1 = }')

r2 = list(range(0,5))   # 증감 부분 생략
print(f'{r2 = }')

r3 = list(range(5))   # 초기값, 증감 부분 생략
print(f'{r3 = }')
```

[심화] - list vs range
- list : 메모리에 모든 값을 넣음

 ## **2.2 for 문 ... range()**

*   **for** 변수 **in** range(start, stop, steps) :  <br> &emsp;문장들


```python
# [실습 7]--------------------------------------
# for ... range() 연습
# 0부터 4차까지 순서대로 출력하기// range(0, 5, 1), range(0, 5), range(1)
# for 문 - range(start=0, stop, steps=1)

for r1 in range(0,5,1) :
    print(f'{r1 }', end=' ')

print()
# for 문 - range(start=0, stop) - 증감 부분 생략
for r2 in range(0,5) :
    print(f'{r2}', end = ' ')

print()
#for 문 - range(stop) - 초기값, 증감 부분 생략
for r3 in range(5) :
    print(f'{r3}', end = ' ')
```

```python
# [실습 8]--------------------------------------
# for ... range() 연습
# 1부터 10까지 순서대로 출력하기
for i in range(1, 11):
    print(i, end=' ')
print()

# 10부터 1까지 역순으로 출력하기
for i in range(10, 0, -1):
    print(i, end=' ')
print()

# 1부터 10까지 홀수만 출력 하기
for i in range(1, 11, 2):
    print(i, end=' ')
print()
```

# **3부.for 누적&필터링 연습**

 ## **3.1 for 누적 예제**
 - 반복문을 돌면서 변수에 값을 계속 더하거나 곱해서 최종 결과를 만듬
-  변수 초기화 (ex. total=0)


```python
# [실습 9]--------------------------------------
# for 누적 연습 - 1에서 10까지 정수의 합 계산

# 1. 누적된 결과를 담을 변수를 0으로 초기화 - 더하기임으로
sum = 0

# 2. 1에서 10까지 반복 - for ... range
for num in range(1, 11) :
    sum += num           #sum = sum + num

print(f'{sum = }')
```

```python
# [실습 10]--------------------------------------
# for 누적 연습 - 1에서 5까지 정수의 곱 계산

# 1. 누적된 결과를 담을 변수를 1로 초기화    - 곱하기임으로

result = 1

# 2. 1에서 5까지 반복 - for ... range
for num in range(1, 6):
    result *= num    #result = result*num

print(f'{result=}')
```

### **[자율1:팩토리얼]**

---
정수 1개를 입력하시오: 5 <br>

```python
# [자율 1]--------------------------------------
# for 누적 연습 - 팩토리얼 계산 ; n! = 1*2*3...*(n-1)*n

# 입력문 - 정수 1개를 입력하시오
number = int(input('정수를 입력하시오: '))

# 초기값
result = 1
# for 문
for num in range(1, number+1):
    result *= num  #result = result*num

print(f'{result=}')
```

 ## **3.2  for 필터링 예제**
 * 여러 개의 데이터 중에서 **조건(if문)을 만족**하는 데이터만 쏙쏙 골라냄
* 골라낸 데이터를 **새로운 리스트**에 담음

```python
# [실습 11]--------------------------------------
# for 필터링 연습 - 리스트에서 짝수만 골라내기

# 1. numbers 리스트 생성 - 1에서 10까지 수
range(1, 11)

# 2. 필터링된 결과(짝수)를 담을 빈 리스트를 만듭니다.
even_numbers = []

for num in range(1, 11):
    if num%2 ==0 :
        even_numbers.append(num)

print(f'{even_numbers=}')

# 홀수
odd_numbers = []

for num in range(1, 11) :
    if num%2 == 1:   # if num%2 != 0:
        odd_numbers.append(num)
print(f'{odd_numbers=}')


odd_numbers = []
for num in range(1, 11, 2) :
    odd_numbers.append(num)
print(f'{odd_numbers=}')
# 3. if문으로 조건을 설정합니다. (2로 나눈 나머지가 0이면 짝수)
```

 ## **3.3  for 누적+필터링 예제**



```python
# [실습 12]--------------------------------------
# for 누적+필터링 연습 - 리스트에서 원하는 것만 골라서 누적 합계

fruits = ['사과', '바나나', '포도', '오렌지', '수박', '파인애플', '복숭아', '키이위']
# 1. 개수를 누적할 변수를 0으로 만듭니다.
count = 0
# 2. for each - fruits

# 3. 글자 수가 3글자인지 확인합니다. (필터링)
for fruit in fruits :
    # print(fruit, len(fruit) )
    if len(fruit) == 3 :
        print(fruit, "당첨! 3글자")
        count += 1  #count = count +1

print(f'{count=}')

```

# **4부. 반복문제어하기(break, continue)**

## **4.1 break : 비상탈출**
* 목표: 특정 조건이 되면 **즉시 반복문을 파괴하고 나옴.**
* 보통 **if 조건문과 같이 사용**

```python
# [실습 13]--------------------------------------
# for 반복문 제어하기 - break

fruits = ['사과', '독사과', '바나나', '포도']

# for each - 독사과이면 비상탈출

for fruit in fruits:
    if fruit == '독사과':
        break

    print(f'{fruit}를 먹었습니다.')
```

## **4.2 continue:건너뛰기**
* 목표: 현재 반복의 **남은 코드는 무시**하고, **다음 순번으로 넘어 감**.
* 보통 **if 조건문과 같이 사용**




```python
# [실습 14]--------------------------------------
# for 반복문 제어하기 - continue

fruits = ['사과', '독사과', '바나나', '포도']

# for each - 독사과이면 건너뛰기

for fruit in fruits:
    if fruit == '독사과':
        continue

    print(f'{fruit}를 먹었습니다.')
```

### **미션 1.  스마트 성적 분석기 (반복문 업그레이드)**

* 5과목의 점수를 `for ~ range`문으로 간편하게 입력받아 리스트에 담습니다. 그 후 `for`문으로 리스트를 순회하며 직접 총점을 누적하고, 80점 이상인 우수 과목만 쏙쏙 필터링하여 개수를 세어봅시다     
* 목표:  for ~range 반복 입력, for 순회(for-each) 누적 합계, if문 데이터 필터링 결합하기


---
**[출력 예시]** <br>

이름을 입력하시오: mju <br>
1번째 과목 정수를 입력하시오: 95<br>
2번째 과목 정수를 입력하시오: 70<br>
3번째 과목 정수를 입력하시오: 85<br>
4번째 과목 정수를 입력하시오: 60<br>
5번째 과목 정수를 입력하시오: 90<br>
-----------------------------------<br>
총점: 400점<br>
평균: 80.0점 (등급: B)<br>
🎉 80점 이상 우수 과목은 총 3개입니다!<br>


----
**[미션 가이드]**

❶  [반복 입력 - for \~ range]  빈 리스트 `scores = []`를 만듭니다. `for i in range(5):`를 사용하여 5과목의 점수를 반복해서 입력받고, `append()`를 이용해 리스트에 차례대로 담으세요.<br>
❷ [누적 합계 & 평균 - for 순회] 총점을 저장할 변수 `total = 0`을 만듭니다. `for score in scores:`를 사용하여 리스트 안의 점수를 하나씩 꺼내며 `total += score`로 합계를 직접 누적합니다. 다 더해졌다면 5로 나누어 평균을 계산하세요.<br>
❸ [등급 판별 - 다중 if문] 계산된 평균 점수를 바탕으로 `if~elif~else`문을 사용하여 등급을 판별합니다. (90점 이상 A, 80점 이상 B, 그 외 C)<br>
❹ [데이터 필터링 - for & if 결합] 우수 과목 개수를 셀 변수 `excellent = 0`을 만듭니다. 리스트를 순회하며(또는 2번 과정에서 함께) `if score >= 80:` 조건을 통해 80점 이상인 점수만 골라내어 카운트(`excellent += 1`) 하세요.<br>


```python
# [미션 1]--------------------------------------
# 스마트 성적 분석기
# 화면 예시에 맞게 먼저 이름을 입력받습니다.
name = input("이름을 입력하시오: ")

# ❶ [반복입력-for~range] 빈 리스트 생성 및 점수 입력
scores = []

for i in range(5) :   # 0, 1, 2, 3, 4
    score = int(input(f"{i+1}번째 과목 정수를 입력하시오: "))
    scores.append(score)


# ❷ [누적합계&평균-for순회] 및 ❹ [데이터 필터링-for&if결합]
total = 0
excellent = 0 # 우수 과목 개수를 셀 변수 초기화

for score in scores:
    total += score # 총점 직접 누적
    if score >= 80: # 80점 이상 우수 과목 카운트
        excellent += 1

# 평균 계산
average = total / len(scores)

# ❸ [등급판별-다중if문] 평균 점수를 바탕으로 등급 판별
if average >= 90:
    grade = 'A'
elif average >= 80:
    grade = 'B'
else:
    grade = 'C'

# 최종 결과 출력
print("-----------------------------------")
print(f"총점: {total}점")
print(f"평균: {average:.1f}점 (등급: {grade})") # 소수점 첫째 자리까지 출력
print(f"🎉 80점 이상 우수 과목은 총 {excellent}개입니다!")
```

```python
# 스마트 성적 분석기 확인 코드
# 1. 테스트를 위해 핵심 로직(누적, 필터링, 등급 판별)을 함수로 만듭니다.
def analyze_scores(scores):
    total = 0
    excellent = 0

    # 누적 합계 및 우수 과목 카운트
    for score in scores:
        total += score
        if score >= 80:
            excellent += 1

    # 평균 계산
    average = total / 5

    # 등급 판별
    if average >= 90:
        grade = 'A'
    elif average >= 80:
        grade = 'B'
    else:
        grade = 'C'

    # 총점, 평균, 등급, 우수 과목 개수를 튜플 형태로 반환합니다.
    return total, average, grade, excellent


# 2. assert 문을 활용한 코드 검증

# 테스트 1: 이미지의 [화면 예시]와 완벽하게 동일한 상황
# 입력: 95, 70, 85, 60, 90
# 기대 결과: 총점 400, 평균 80.0, 등급 'B', 우수 과목 3개
assert analyze_scores([95, 70, 85, 60, 90]) == (400, 80.0, 'B', 3), "테스트 1 실패: 화면 예시 결과와 다릅니다."

# 테스트 2: 모든 과목이 90점 이상인 최우수 학생 (A등급)
# 입력: 90, 100, 95, 100, 90
# 기대 결과: 총점 475, 평균 95.0, 등급 'A', 우수 과목 5개
assert analyze_scores([90, 100, 95, 100, 90]) == (475, 95.0, 'A', 5), "테스트 2 실패: A등급 판별 및 전체 우수 과목 카운트 오류"

# 테스트 3: 모든 과목이 80점 미만인 상황 (C등급)
# 입력: 70, 65, 50, 75, 60
# 기대 결과: 총점 320, 평균 64.0, 등급 'C', 우수 과목 0개
assert analyze_scores([70, 65, 50, 75, 60]) == (320, 64.0, 'C', 0), "테스트 3 실패: C등급 판별 및 우수 과목 0개 카운트 오류"

# 테스트 4: 평균이 정확히 90.0인 경계값 테스트 (A등급 턱걸이)
# 입력: 80, 100, 90, 90, 90
# 기대 결과: 총점 450, 평균 90.0, 등급 'A', 우수 과목 5개
assert analyze_scores([80, 100, 90, 90, 90]) == (450, 90.0, 'A', 5), "테스트 4 실패: 평균 90점 정각(경계값) A등급 판별 오류"

# 위 assert 문들을 모두 통과했다면 아래 메시지가 출력됩니다.
print("✅ 미션 1: 스마트 성적 분석기 로직 검증을 완벽하게 통과했습니다!")
```

❶   [초기화]   총 누적 금액을 저장할 변수 `total_money`를 `0`으로 만듬<br>
❷   [반복문 설정]   for를 사용 하여 1일부터 5일까지 반복하는 뼈대를 만듬<br>
❸   [입력 및 필터링 - continue] 입력된 금액이 0보다 작으면  -  “잘못된 금액입니다”를 출력하고,   continue를 사용하여   누적하지 않고 넘어감<br>
❹   [금액 누적 및 확인]   정상적인 금액이라면 `total_money`에 누적(`+=`)하고, 현재까지 모인 금액을 화면에 출력.<br>
❺   [목표 달성 확인 - break]  total_money가 10000원 이상인지 확인.하고,   10000원 이상이라면 "🎉 목표 금액(10000원)을 조기 달성했습니다"를 출력하고,  break를 사용하여 반복문을 즉시 탈출.<br>
❻   [종료 안내]   `for`문이 끝난 다음에, 5일이 무사히 지나 목표에 미달했을 경우를 대비하여 ＂최종 모은 금액은 OOO원입니다.＂를 출력하는 코드를 작성<br>



```python
# [미션 2]--------------------------------------
# 스마트 저금통 : 목표 금액 달성 프로젝트
# ❶ [초기화] 총 누적 금액을 저장할 변수 0으로 만들기
# 누적 금액 변수 - total_money
total_money = 0

# ❷ [반복문 설정] 1일부터 5일까지 반복
for day in range(1, 6):
    deposit = int(input(f"[ {day} 일차 ] 저금할 금액 : "))

    # ❸ [입력 및 필터링 - continue] 음수 입력 시 건너뛰기
    if deposit < 0:
        print("잘못된 금액입니다. (건너뜀)")
        continue

    # ❹ [금액 누적 및 확인] 정상적인 금액 누적 및 출력
    total_money += deposit
    print(f"현재 누적 금액 : {total_money}원")

    # ❺ [목표 달성 확인 - break] 누적 금액이 10000원 이상인지 확인
    if total_money >= 10000:
        print("-------------------------------------------")
        print("🎉 축하합니다! 목표 금액(10000원)을 조기 달성했습니다!")
        break

# ❻ [종료 안내] 5일이 무사히 지나 목표에 미달했을 경우 출력
else: # for 루프가 break 없이 정상적으로 완료되었을 때 실행
    print("------------------------------------------------")
    print(f"저금 기간이 끝났습니다. 최종 모은 금액은 {total_money}원입니다.")
```

```python
# 1. 테스트를 위해 핵심 로직을 함수로 만듭니다.
def simulate_piggy_bank(deposit_list):
    total_money = 0

    # 전달받은 리스트에서 최대 5일치(5번)만 반복합니다.
    for amount in deposit_list[:5]:
        # 음수면 건너뛰기
        if amount < 0:
            continue

        # 금액 누적
        total_money += amount

        # 10,000원 이상이면 즉시 탈출
        if total_money >= 10000:
            break

    return total_money

# 2. assert 문을 활용한 코드 검증 (결과가 참이면 아무것도 출력되지 않고 넘어갑니다)

# 테스트 1: 이미지의 [화면 예시 1]과 같은 상황 (조기 달성 및 음수 예외 처리)
# 5000원 + (-1000원 무시) + 6000원 = 11000원 (여기서 달성하고 break 되어야 함)
assert simulate_piggy_bank([5000, -1000, 6000, 2000, 3000]) == 11000, "테스트 1 실패: 조기 달성 및 예외 처리 오류"

# 테스트 2: 5일이 지나도 목표를 달성하지 못한 상황
# 1000 + 2000 + (-500 무시) + 3000 + 1000 = 7000원
assert simulate_piggy_bank([1000, 2000, -500, 3000, 1000]) == 7000, "테스트 2 실패: 5일 미달성 누적 금액 오류"

# 테스트 3: 1일차에 바로 목표 금액을 달성한 상황
# 15000원에서 바로 break 되어야 함 (뒤의 금액은 더해지지 않음)
assert simulate_piggy_bank([15000, 2000, 3000]) == 15000, "테스트 3 실패: 1일차 조기 종료 오류"

# 테스트 4: 음수만 계속 입력된 상황
assert simulate_piggy_bank([-1000, -2000, -3000, -4000, -5000]) == 0, "테스트 4 실패: 음수 누적 오류"

# 위 assert 문들을 모두 무사히 통과했다면 아래 메시지가 출력됩니다.
print("✅ 모든 Assertion 테스트를 완벽하게 통과했습니다! 코드가 잘 짜였습니다.")
```
