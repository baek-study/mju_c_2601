### **[예제2] 구구단** <br>
2단 부터 9단 까지 각각
1부터 9까지 곱으로 표현

```python
# [실습 11]--------------------------------------
# 중첩 for 예제 -2단 부터 9단까지 1부터 9까지 곱으로

for dan in range(2, 10) : # 2,3,4,5,6,7,8,9,
    for gob in range(1, 10) : # 1,2,3,4,5,6,7,8,9
        print(f'{dan}*{gob}={dan*gob}')
    print()
```

####[심화]2단 부터 9단까지 1부터 9까지 곱으로 - 각 단의 곱을 한줄에 표현

```python
# [심화]--------------------------------------
# 중첩 for 예제 - 2단 부터 9단까지 1부터 9까지 곱으로 - 각 단의 곱을 한줄에 표현

for dan in range(2, 10):
    for gob in range(1, 10):
        print(f'{dan}*{gob}={dan*gob}', end='\t') # Use tab for separation
    print() # Move to the next line after each dan
```

### **[예제3] 주사위 pair 찾기**
- 주사위 2개(dice1, dice2)의 합이 6이 되는 pair 출력
- 전체탐색(brute force) : 모든 경우의 수 (무식하게) 검색 - 비효율적이나 정확함

```python
# [실습 12]--------------------------------------
# 중첩 for 예제 - # 주사위 pair 칮기

for dice1 in range(1,7) :
    for dice2 in range(1, 7) :
        if dice1+dice2 == 6:
            print(f'{dice1}+{dice2} = {dice1+dice2}')
```

### **미션 1. 나만의미니카페키오스크**

* 무한 루프를 사용하여 종료를 선택할 때까지 계속해서 주문을 받는 키오스크를 만듭니다. 선택한 메뉴에 따라 총 결제 금액을 누적하고, 잘못된 번호를 입력하면 예외 처리를 해봅시다.  
* 목표:  while True 무한 루프 이해, += 연산자로 변수 누적하기, break와 continue를 활용한 제어




---
**[출력 예시]** <br>

=== ☕ 미니 카페 ===<br>
1. 아메리카노 (2000원)<br>
2. 카페라떼 (3000원)<br>
3. 주문 완료 (종료)<br>

메뉴 선택: 1<br>
👉 아메리카노 추가! (현재 금액: 2000원)<br>

메뉴 선택: 5<br>
❌ 없는 메뉴입니다. 다시 선택해주세요.<br>

메뉴 선택: 2<br>
👉 카페라떼 추가! (현재 금액: 5000원)<br>

메뉴 선택: 3<br>
🎉 주문 완료! 총 결제 금액은 5000원입니다.<br>


----
**[미션 가이드]**

❶  [변수 준비 & 무한 루프]  총 금액을 저장할 변수 total_price = 0을 만듭니다. 그 아래에 while True:를 사용하여 메뉴판을 출력하고 번호를 입력받는 무한 루프를 시작하세요.<br>
❷  [조건 판별 & 금액 누적]  if~elif 문을 사용하여 1번(아메리카노)이면 2000원, 2번(카페라떼)이면 3000원을 total_price에 직접 누적(+=)합니다.<br> 메뉴에 없는 번호를 눌렀다면 continue를 써서 다시 처음(메뉴판)으로 돌아가게 만드세요.<br>
❸  [주문 종료 & break] 사용자가 3번(주문 완료)을 선택하면 누적된 총 결제 금액을 출력하고, break를 사용하여 무한 루프를 완전히 빠져나옵니다


```python
# 미션
# 1. 총 금액을 누적할 변수 준비
total_price = 0

# 2. 무한 루프
while True:
    # 3. 메뉴 쓰기
    print('\n=== ☕ 미니 카페 ===')
    print('1. 아메리카노 (2000원)')
    print('2. 카페라떼 (3000원)')
    print('3. 주문 완료 (종료)')

    try:
        # 4. 메뉴 입력
        menu = int(input('메뉴 선택: '))
    except ValueError:
        print('❌ 숫자를 입력해주세요.')
        continue

    # 5. 연속 if - 메뉴 선택
    if menu == 1 :
        total_price += 2000
        print(f'👉 아메리카노 추가! (현재 금액: {total_price}원)')
    elif menu == 2:
        total_price += 3000
        print(f'👉 카페라떼 추가! (현재 금액: {total_price}원)')
    elif menu == 3:
        print(f'🎉 주문 완료! 총 결제 금액은 {total_price}원입니다.')
        break
    else :
        print('❌ 없는 메뉴입니다. 다시 선택해주세요.')
```

```python
import random

# 1에서 100 사이의 임의의 숫자 생성
target_number = random.randint(1, 100)
attempts = 0

print("1에서 100 사이의 숫자를 추측 해 보세요 :")

while True:
    try:
        guess = int(input("숫자를 입력하세요: "))
        attempts += 1

        if guess < target_number:
            print("정답은 더 큽니다.")
        elif guess > target_number:
            print("정답은 더 작습니다.")
        else:
            print(f"정답입니다. {attempts}번만에 맞추셨군요!")
            break
    except ValueError:
        print("유효한 숫자를 입력해주세요.")
```

```python
# 실습문제 4
# 사용자에게 두 개의 정수 a와 b를 입력 받고, 1부터 500이하의 정수 중
# a의 배수면서 b의 배수인 수를 출력하는 프로그램 작성.
# 단, 사용자가 1~500사이의 두 개의 정수를 정확하게 입력할 때 까지 입력을 반복 수행

while True:
    a = int(input("첫 번째 정수를 입력하시오 (1-500): "))
    b = int(input("두 번째 정수를 입력하시오 (1-500): "))
    if 1 <= a <= 500 and 1 <= b <= 500 :
        break
    else:
        print("1에서 500 사이의 정수를 입력해주세요.")


print(f"1부터 500까지 {a}와 {b}의 공배수:")
for i in range(1, 501):
    if i % a == 0 and i % b == 0:
        print(i)
```
