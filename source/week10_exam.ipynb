## **1.3 문자열 처리**

* 인덱싱 : 개별 문자 반환, msg[1]
* 슬라이싱 : 부분 문자열 반환, msg[2:4]


### **1) 인덱싱 : 문자열[index]**
- 특정위치의 개별 문자 반환

```python
## 1. 인덱싱 : [index] 특정위치 반환 - 0 부터 시작
msg = "안녕하세요"

# msg의 0 번째 인덱스 출력
print(msg[0])
# msg의 -1번째 인덱스 출력
print(msg[1])
# msg에서 '하' 를 출력 - 양의 인덱스, 음의 인덱스 이용
print(msg[2])
print(msg[-3])
```

### **2) 슬라이싱 : 문자열[시작:종료]**
* **기본** : 문자열[시작:종료] - 시작부터 종료 직전(종료-1) 까지 부분 문자열 반환
* **시작 생략** : 문자열[:종료] -  첫문자 부터
* **종료 생락** : 문자열[시작:] -  마지막 문자 까지

```python
## 2) 슬라이싱1 : [시작:종료] 시작부터 끝 직전(끝-1) 위치까지 부분 문자열 반환
msg = '안녕하세요'


# 문자열 '하세'  출력  - 2, 3
print(msg[2:4])

# 문자열 '녕하세' 출력 - 1, 2, 3
print(msg[1:4])

# 문자열 '하세요' 출력 - 2, 3, 4
print(msg[2:5])

# 문자열 '안녕' 출력  - 0, 1
print(msg[0:2])

# 문자열 시작 생략 가능 : 문자열 첫 문자부터
print(msg[:2])

# 문자열 끝 생략 : 문자열 마지막 문자까지
print(msg[2:])
```

```python
## 2) 슬라이싱2 : 문자열을 2부분으로 나누기

msg = "20260505Sunny"  # 인덱스 8 이후

# date 0~7
date = msg[0:8]
date1 = msg[:8]

# weather
weather = msg[8:13]
weather1 = msg[8:]

print(f'{date=}, {weather=}')
print(f'{date1=}, {weather1=}')
# year 0~3
year = date[:4]
# month 4~5
month = date[4:6]
# day 6~7
day = date[6:]

print(f'{date=}, {year=}, {month=}, {day=}')
```

## **2.3 f-문자열:자료형식**
{표현식:자료형식코드}

*   정수 : d(decimal), {num:d}
*   실수 : f(fixed/float), {avg:f}
* 문자열 : s (string), {msg:s}



### **1) 정수**
f'.... {표현식:**d**} ....'  
- d(decimal, 10진수)
- b(binary, 2진수)
- x(hexa, 16진수)

```python
num = 15

print(f'2진 표현으로는 {num:b}입니다.')
print(f'10진 표현으로는 {num:d}입니다.')
print(f'16진 표현으로는 {num:x}입니다.')
```

### **2) 자릿수&정렬**
f'.... {표현식:**5d**} ....'  
- 5d : 다섯자리 확보
- <5d, ^5d: 왼쪽 정렬, 가운데 정렬
- 05d : 다섯자리에 0 채우기

```python
num = 15

print(f'[{num:d}]')           # 기본 변환
print(f'[{num:5d}]')           # 다섯 자리 확보
print(f'[{num:<5d}]')           # 왼쪽 정렬
print(f'[{num:^5d}]')           # 가운데 정렬
print(f'[{num:05d}]')           # 0 채우기
print(f'[{12345678:5d}]')      # 모자라면 지정된 자릿수 무시
```

### **3) 실수**
f'.... {표현식:**f**} ....'  
- f: 소수점 표기 형식
- e: 과학적(지수) 표기 형식

```python
num = 15

# num 출력 - 10진 표현

n1 = 1000
n2 = 8

avg = n1/n2

print(f'{n1} / {n2} = {avg}')
print(f'{n1} / {n2} = {avg:f}')    # 실수 f 형식
print(f'{n1} / {n2} = {avg:e}')    # 실수 e 형식
```

### **4)정밀도:소수점 몇자리**


* {PI:.2f} # 소수점 이하 두자리
* {PI:5.2f} # 전체가 5자리(소수점 포함)



```python
PI = 3.14159265

print(f'[{PI:f}]')            # 기본적으로 소수점 이하 여섯 자리
print(f'[{PI:.2f}]')          # 소수점 이하 두 자리
print(f'[{PI:5.2f}]')         # 소수점 기호 .도 한 자리를 차지
```

### **3) 자르기/합하기: split()/join()**
- 문자열.split(구분자) : 여러 문자열로 분리
- 구분자.join(리스트) : 여러 문자열을 합치기


```python
# 날짜 기호 바꾸기 ( / 에서 - 로 )
date_input = "2026/05/05"
print("원본 날짜:", date_input)

# [split]: 슬래시('/')를 기준으로 년, 월, 일을 분리
date_parts = date_input.split('/')
print("분리된 데이터:", date_parts)

date_input1 = '2026++05++05'
list1 = date_input1.split('++')
print(f'{list1=}')

# [join]: 분리된 데이터 사이에 하이픈('-')을 넣어서 조립
clean_date = '-'.join(date_parts)
print("수정된 날짜:", clean_date)
```

### **미션 1. 성적표 만들기**

* 사용자로부터 4명 학생의 성적 정보(이름, 과목 점수)를 한 줄로 입력받아, 형변환 후 1차원 리스트로 만들고, 이를 모아 전체 성적표(2차원 리스트)를 완성해 봅시다.   
* 목표: input().split()을 활용한 다중 입력 처리, int() 형변환, 1차원/2차원 리스트 구조 이해 및 append() 활용



---
**[출력 예시]** <br>

학생 4명의 정보(이름 파이썬 국어)를 띄어쓰기로 입력<br>

1번 학생: 마루 95 80<br>
2번 학생: 명지 88 90<br>
3번 학생: 박변수 70 85<br>
4번 학생: 최함수 100 95<br>
<br>
--- 📋 전체 성적표 ---<br>
' [ [＇마루', 95, 80], [＇명지', 88, 90],['박변수', 70, 85], ['최함수', 100, 95] ] '

----
**[미션 가이드]**

❶  **[빈 리스트 준비 & 반복문]**  전체 학생의 성적을 담을 빈 2차원 리스트 grade_book = []을 만듭니다. 그 아래에 4명의 학생 정보를 받기 위해 for문을 사용하여 4번 반복하는 루프를 만드세요.<br>
❷ **[입력 및 데이터 분해]**   input()과 split()을 함께 사용하여 이름과 2과목(파이썬, 국어,) 점수를 띄어쓰기 기준으로 한 번에 입력받고 쪼개어 변수에 저장합니다.<br>
❸  **[형변환 및 2차원 리스트 추가]** <br>
- 쪼개진 데이터 중 이름은 그대로 두고,  점수는 int()로 숫자로 변환<br>
- 변환된 4개의 데이터로 1차원 리스트(한 학생의 성적)를 만든 뒤, grade_book.append()를 사용해 2차원 리스트에 추가합니다. <br>

➍ **[2차원 리스트 출력]**  print를 이용하여 최종 성적표를 출력하세요<br>


```python
# 미션1

# 1. 빈 2차원 리스트 준비
grade_book = [ ]

# 2. 반복문 - 4명의 학생 정보를 받기
for _ in range(0, 4):
    # 3. 데이터 입력 및 분해
    name, python, kor = input("학생 4명의 정보(이름 파이썬 국어)를 띄어쓰기로 입력: ").split()

    # 4. 형변환
    python = int(python)
    kor = int(kor)

    # 5. 변환된 데이터로 1차원 리스트 만들기
    student = [name, python, kor]

    # 6. 2차원 리스트(grade_book)에 조립한 1차원 리스트 추가하기
    grade_book.append(student)

# 7. 2차원 리스트 출력
print("\n--- 📋 전체 성적표 ---")
print(grade_book)
```

### **미션 2. 성적표 만들기(심화)**

* 미션 1에서 만든 코드를 활용하여, 각 학생의 '총점'과 '평균'을 컴퓨터가 직접 계산하게 만들고, 성적표를 보기 좋게 예쁘게 출력해 봅시다!  
* 목표: 산술 연산자(+, /) 활용, 리스트 데이터 구조 확장, for문과 f-string을 활용한 데이터 순회 및 포매팅

---
**[출력 예시]** <br>

학생 4명의 정보(이름 파이썬 국어)를 띄어쓰기로 입력<br>

1번 학생: 마루 95 80<br>
2번 학생: 명지 88 90<br>
3번 학생: 박변수 70 85<br>
4번 학생: 최함수 100 95<br>

--- 📊 최종 성적 분석 결과 ---<br>
마루 학생 ➡️ 파이썬: 95, 국어: 80 | 총점: 175, 평균: 87.5<br>
명지 학생 ➡️ 파이썬: 88, 국어: 90 | 총점: 178, 평균: 89.0<br>
박변수 학생 ➡️ 파이썬: 70, 국어: 85 | 총점: 155, 평균: 77.5<br>
최함수 학생 ➡️ 파이썬: 100, 국어: 95 | 총점: 195, 평균: 97.5<br>


----
**[미션 가이드]**

❶  **[데이터 입력 & 형변환]**  미션 1의 코드를 그대로 활용합니다. (리스트 grade_book 만들고, 4번 반복 (이름, 2과목 입력 받고 과목은 정수변환)<br>
❷ **[총점 및 평균 계산 추가]**    변환된 파이썬 점수와 국어 점수를 더해 총점을 구하고, 그 총점을 2로 나누어 평균을 구하는 변수를 만드세요.<br>
❸  **[1차원 리스트 조립 & 누적]**   미션 1과 달리 이번에는 [이름, 파이썬, 국어, 총점, 평균] 순서로 5개의 데이터가 담기도록 1차원 리스트를 조립하고 grade_book에 추가(append)합니다.<br>
➍ **[최종 성적표 예쁘게 출력]**  데이터 입력 루프가 모두 끝난 후,   새로운 for문과 f-string을 사용하여 2차원 리스트(grade_book)에 담긴 학생들의 성적을 화면 예시처럼 보기 좋게 출력해 보세요.  (평균을 소수점2자리로 출력 하시기 바랍니다)<br>

```python
# 미션2

# 1. 빈 2차원 리스트 준비
grade_book = []

# 2. 반복문 - 4명의 학생 정보를 받기
for _ in range(4) :
    # 3. 데이터 입력 및 분해
    name, python, kor = input("학생 4명의 정보(이름 파이썬 국어)를 띄어쓰기로 입력: ").split()

    # 4. 데이터 할당 및 형변환 (점수는 int로!)
    python = int(python)
    kor = int(kor)

    # 5. 🌟 추가된 부분: 파이썬에게 계산 시키기!
    total = python + kor
    avg = total/2

    # 6. 1차원 리스트 조립 (미션에서 요구한 5개 순서대로)
    student=[name, python, kor, total, avg]

    # 7. 완성된 1차원 리스트를 2차원 리스트에 누적
    grade_book.append(student)  # grade_book.append([name, python, kor, total, avg])

# 🌟 추가된 부분: for문과 f-string으로 예쁘게 출력하기
print("\n--- 📊 최종 성적 분석 결과 ---")
# 반복문..
for student in grade_book :
    #print(student)
    print(f'{student[0]} 학생 ➡️ 파이썬: {student[1]}, 국어: {student[2]} | 총점: {student[3]}, 평균: {student[4]}')
    # student는 [이름, 파이썬, 국어, 총점, 평균] 형태의 리스트입니다.
    # 인덱스 번호를 활용하여 f-string에 쏙쏙 집어넣습니다.
```

### **미션 3: 나만의 K-POP 아이돌 데뷔조 결성!(챌린지)**

* 3명의 연습생 정보(이름, 보컬, 댄스)를 입력받아 평균 점수를 계산합니다. 만약 평균이 85점 이상이라면 '합격'시켜 데뷔조(2차원 리스트)에 추가하고, 아니라면 탈락시킵니다. 최종 데뷔 멤버를 확인해 봅시다  
* 목표: 데이터 분해(split), 연산, if 조건문을 활용한 데이터 필터링, 2차원 리스트 종합 활용

---
**[출력 예시]** <br>

[ 오디션 심사 시작 ] 연습생 정보(이름 보컬 댄스) 입력<br>
>> 지원자 1: 마루 90 95<br>
🎉 마루 데뷔조 합격! (평균: 92.5점)<br>
<br>
>> 지원자 2: 자니 80 80<br>
❌ 자니 탈락입니다. (평균: 80.0점)<br>
<br>
>> 지원자 3: 명지 100 85<br>
🎉 명지 데뷔조 합격! (평균: 92.5점)<br>
-----------------------------------<br>
🌟 최종 데뷔조 명단 🌟<br>
[[＇마루', 92.5], [＇명지', 92.5]]<br>



----
**[미션 가이드]**

3명의 연습생 정보(이름, 보컬, 댄스)를 입력받아 평균 점수를 계산합니다. 만약 평균이 85점 이상이라면 '합격'시켜 데뷔조(2차원 리스트)에 추가하고, 아니라면 탈락시킵니다. 최종 데뷔 멤버를 확인해 봅시다
-  목표: 데이터 분해(split), 연산, if 조건문을 활용한 데이터 필터링, 2차원 리스트 종합 활용

❶  **[초기화]**  합격한 멤버들을 모아둘 빈 2차원 리스트 debut_group  만듬<br>
❷ **[반복문 & 데이터 입력]**   for를 사용하여 3번 반복하는 뼈대를 만들고, input().split()으로 이름, 보컬, 댄스 점수를 한 번에 입력받아 변수에 저장<br>
❸  **[형변환 및 평균 계산]**  보컬과 댄스 점수는 정수(int)로 바꾸고, 두 점수를 더해 2로 나눈 평균 변수를 만듬<br>
➍ **[조건 판별 & 데뷔조 합류]**   if문을 사용하여 평균이 85점 이상인지 확인<br>
-  85점 이상이라면: 화면에 "🎉 [이름] 데뷔조 합격!"을 출력하고, [이름, 평균] 형태의 1차원 리스트로 만들어 debut_group에 추가<br>
-  85점 미만이라면: "❌ [이름] 탈락입니다."를 출력 <br>
➎  **[최종 결과 발표]**  반복문이 모두 끝나면, 2차원 리스트(debut_group)를 활용해 최종 데뷔조 명단을 멋지게 출력해 보세요!<br>

```python
# 챌린지

# ❶ 초기화: 합격한 멤버들을 모아둘 빈 2차원 리스트 debut_group 생성
debut_group = []

print("[ 오디션 심사 시작 ] 연습생 정보(이름 보컬 댄스) 입력")
# ❷ 반복문 & 데이터 입력: for를 사용하여 3번 반복하는 뼈대를 만들고,
# input().split()으로 이름, 보컬, 댄스 점수를 한 번에 입력받아 변수에 저장
for i in range(1, 4):
    data = input(f">> 지원자 {i}: ").split()
    name = data[0]
    vocal = int(data[1])
    dance = int(data[2])

    # ❸ 형변환 및 평균 계산: 보컬과 댄스 점수는 정수(int)로 바꾸고,
    # 두 점수를 더해 2로 나눈 평균 변수를 만듬
    average = (vocal + dance) / 2

    # ❹ 조건 판별 & 데뷔조 합류: if문을 사용하여 평균이 85점 이상인지 확인
    if average >= 85:
        # 85점 이상이라면: 화면에 "🎉 [이름] 데뷔조 합격! (평균: [평균]점)"을 출력하고,
        # [이름, 평균] 형태의 1차원 리스트로 만들어 debut_group에 추가
        print(f"🎉 {name} 데뷔조 합격! (평균: {average:.1f}점)")
        debut_group.append([name, average])
    else:
        # 85점 미만이라면: "❌ [이름] 탈락입니다."를 출력
        print(f"❌ {name} 탈락입니다. (평균: {average:.1f}점)")
    print()

# ➎ 최종 결과 발표: 반복문이 모두 끝나면, 2차원 리스트(debut_group)를 활용해
# 최종 데뷔조 명단을 멋지게 출력해 보세요!
print("-----------------------------------")
print("🌟 최종 데뷔조 명단 🌟")
print(debut_group)
```
